```python
import ast
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(ast.literal_eval)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
def convert_cast(obj):
    count=0
    L=[]
    for i in ast.literal_eval(obj):
        if count<3:
            L.append(i['name'])
            count+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
def convert_crew(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.iloc[0].overview
```

```python
movies.head(75)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: [i.replace(" ","") for i in x])
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movies = movies[['movie_id', 'title', 'tags']]
```

```python
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies['tags'] = movies['tags'].apply(lambda x: x.lower())
```

```python
movies['tags'][0]
```

```python
movies.duplicated().sum()
```

```python
movies.drop_duplicates(inplace=True)
```

```python
movies.shape
```

# Vectorization 
Bag of Words technique


```python
movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

# Cosine Similarity
--- minimum distance between two vectors means how much similarity between two objects 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

# recommender function

```python
movies
```

```python
def recommend(movie):
    
    movie_index = int(movies[movies['title'] == movie].index[0])
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)), reverse = True, key = lambda x:x[1])[1:11]

    for i in movie_list:
        print(movies.iloc[i[0]]['title'])
    
```
